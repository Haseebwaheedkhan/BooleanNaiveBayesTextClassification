Remove Duplicate Words from a document

```python
def RemoveDuplicate(lines):
    '''
    Argument : String 
    Return   : String
    
    Take input as string and remove duplicate words from the string
    '''
    temp_word = ""
    line_without_duplication = ""
    for char in lines:
        if char == " " and len(temp_word)!=0:
            if temp_word not in line_without_duplication:
                line_without_duplication = line_without_duplication+ char + temp_word
            temp_word = ""
        else:
            temp_word+=char
    return line_without_duplication
```

Reading Files

```python
def Read_file(folder_path):
    '''
    Argument : 
    Folder_path = Provide complete path of folder where all other text documents are placed

    Returns: 
    Return a complete string copying data from all indivisual files placed in the same folder
    '''

    # Reteriving all file names available in Real folder
    file_name = os.listdir(folder_path)
    # Save all text from all files avaliable in one folder to make a complete document in the form of string
    all_lines = "" 
    for name in file_name:
        file = open(name)
        line = file.read().replace("\n", " ")
        #Removing Duplicate words from a sentence
        line = RemoveDuplicate(line)
        # print(name,"\n",line)
        file.close()
        all_lines+=line
    return all_lines , len(file_name)
 
read = Read_file(path)
```

Word Tokenization

```python
def Tokenization(all_lines):
    '''
    Input : String of complete document
    Return : list of words
    
    Function takes input as a string and tokenize words after formating it into 
    desired shape
    '''
    tokenize_words = list()
    remove_char = ["۔","،","(",")","،","'","`","’",":",",","‘",".","—","ء","“","ء","”",";","—"]
    
    # Tokenize words using NLTK 
    for w in word_tokenize(all_lines):
        temp_word = ""
        # Check each char in word and remove punctuation marks from word
        for char in w:
            if char not in remove_char:
                temp_word+=char
        # Append word in list if word is not empty
        if len(temp_word) != 0:
            tokenize_words.append(temp_word) 
            # print(temp_word, "    ",w)
    return tokenize_words

# all_tokenize_word = Tokenization(real_data)
```

Finding Frequency of word in a Document

```python
def word_frequency(list_of_words):
    '''
    Argument : List of tokenize words in the form of List
    Return : Dictionary of words as key and frequency as values
    '''
    word_dic = dict()

    for word in list_of_words:
        if word not in word_dic:
            word_dic[word] = 1
        else:
            word_dic[word] += 1
    return word_dic

word_frequency(train_real_word)
```

Calculating probility of prior for both Classes i.e Fake & Real. for Trainining

```python
def calculate_prob(unique_words,all_class_words,word_with_frequency):
    word_prob = dict()
    for word in unique_words:
        if word not in word_with_frequency:
            word_prob[word] = 1/( len(all_class_words) + len(unique_words) )
        else:
            word_prob[word] = ( word_with_frequency[word]+1 )/( len(all_class_words) + len(unique_words) )
    return word_prob
```

```python
def BooleanNaiveBase(train_real_word, train_fake_word,len_tr_real,len_tr_fake):

    # (train_real_word, train_fake_word, test_real_word, test_fake_word,len_tr_real,len_tr_fake,len_ts_real,len_ts_fake) = Get_data()
    prior_real = len_tr_real/(len_tr_real+len_tr_fake)
    prior_fake = len_tr_fake/(len_tr_real+len_tr_fake)
    # Calculating words freq 
    real_word = word_frequency(train_real_word)
    fake_word = word_frequency(train_fake_word)
    # Extracting Unique words from complete document
    real_word_keys = list(real_word.keys()) # Unique words in Real Dataset
    fake_word_keys = list(fake_word.keys()) # Unique words in Fake Dataset
    unique_words = [word for word in real_word_keys]
    unique_words = unique_words + [word for word in fake_word_keys]
    
    #Calculating probilities for Real & Fake Class 
    prob_real = calculate_prob(unique_words,train_real_word,real_word)
    prob_fake = calculate_prob(unique_words,train_fake_word,fake_word)

    return (prob_real, prob_fake, prior_real, prior_fake, unique_words)

```

# Performance Evaluation

```python
def ShowPerfomanceEvaluation(final_result):

    TP = sum(final_result['Real'])
    TN = sum(final_result['Fake'])
    FN = len(final_result['Fake'])-sum(final_result['Fake'])
    FP = len(final_result['Real'])-sum(final_result['Real'])

    # Calculating Perfomance
    Accuracy = (TP+TN)/(TP+TN+FP+FN)
    Precision = TP/(TP+FP)
    Recall = TP/(TP+FN)
    F = (2*Precision*Recall) / (Precision+Recall)

    # Print Model Perfomance in the form of a table
    x = PrettyTable()
    x.field_names = ["Sr No.","Perfomance Evaluator", "Value"]

    x.add_row(["01","F", F])
    x.add_row(["--","--------", "------------------"])
    x.add_row(["02","Recall", Recall])
    x.add_row(["--","--------", "------------------"])
    x.add_row(["03","Accuracy", Accuracy])
    x.add_row(["--","--------", "------------------"])
    x.add_row(["04","Precision", Precision])


    print(x)
```
